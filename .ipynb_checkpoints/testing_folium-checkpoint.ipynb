```python
import folium 

# Create a map centered at a specific latitude and longitude, the first parameter is the lat and long, the second 
# is the zoom value: 1 will show you the whole map, 5 will show you a Country and every thing that surrounds it and so on
m = folium.Map(location=[6.1811, 35.7469], zoom_start=5)

# Add a marker, the first parameter puts a marker in the lat and long of the the map, the second popup is the information shown when clicked
# and the third is the tooltip which appears when the mouse hovers over the icon 
folium.Marker([6.1811, 35.7469], popup="Dodoma", tooltip="Click me!").add_to(m)

# Display the map,  this is the way you show the map you are creating  
m
```

```python
# This code creates and shows mulitple positions on a map from a dictonary 

# Sample locations (City, Latitude, Longitude)
locations = [
    ("Dodoma", -6.161184, 35.7469),
    ("Dar es Salaam", -6.776012, 39.2894),
    ("Zanzibar", -6.16394, 39.19793),
    ("Morogoro", -6.830373, 37.670589) ]

# Creates a map
m = folium.Map(location=[-6.2217, 34.8888], zoom_start=4)  # Center of Tanzania

# Adds markers for each location
for city, lat, lon in locations:
    folium.Marker(
        location=[lat, lon],
        popup=city,
        tooltip=f"Click for {city}"
    ).add_to(m)

# Displaying the created map  
m
# This code is used to generated
```

```python
# This code is used to generated a circle around a postition on the map 


folium.CircleMarker(
    location=[-6.2217,  34.8888],
    radius=10,  # Size of the circle
    color="red",
    fill=True,
    fill_color="red"
).add_to(m)

m
```
